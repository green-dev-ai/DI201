```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('US Superstore data.xls')

#train.dropna(subset=['Embarked'], inplace=True)

data.head()
data.info()
```

```python
order_dates = data['Order Date']
Sales = data['Sales']

data["Order Date"] = pd.to_datetime(data["Order Date"])
data["Year"] = data["Order Date"].dt.year

sales_by_year = data.groupby('Year')['Sales'].sum()

sales_by_year.plot(kind="line")

plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
states = data['State']
Sales = data['Sales']

sales_by_state = data.groupby('State')['Sales'].sum()

sales_by_state.plot(kind="line")

plt.title("Sales by State")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.show()
```
