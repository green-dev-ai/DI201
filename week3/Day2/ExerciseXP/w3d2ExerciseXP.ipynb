```python
# --------- Excersice 1 - Duplicates
import pandas as pd
train = pd.read_csv('train.csv')
train_no_duplicates = train.drop_duplicates()
if len(train_no_duplicates) < len(train):
    print(f"There were {len(train) - len(train_no_duplicates)} duplicate lines that were removed from the data")
else:
    print(f"There were no duplicated to remove")

if True in train.duplicated():
    print(f"duplicates")
else:
    print(f"no duplicates")
```

```python
# -------------- Excersice 2 - Missing values
from sklearn.impute import SimpleImputer

# remove lines with no value at Embarked

train["Cabin"] = train["Cabin"].fillna("No Cabin")
train.dropna(subset=['Embarked'], inplace=True)

# replace lines with no Age value with the mean Age

imputer = SimpleImputer(strategy="mean")
train["Age"] = imputer.fit_transform(train[["Age"]])
```

```python
# ---------- Exercise 3 

train["Family_Size"] = 1 + train["SibSp"] + train["Parch"]

train["Title"] = train["Name"].str.extract(r",\s*([^\.]+)\.")

train
```

```python
# ---------- Exercise 4

# Identify and handle outliers in 'Fare' column
P2 = train['Fare'].quantile(0.02)
P98 = train['Fare'].quantile(0.98)
IPR = P98 - P2
lower_bound = P2 - 1.5 * IPR
upper_bound = P98 + 1.5 * IPR
train = train[(train['Fare'] >= lower_bound) & (train['Fare'] <= upper_bound)]

len(train)

# Identify and handle outliers in 'Age' column
P2 = train['Age'].quantile(0.02)
P98 = train['Age'].quantile(0.98)
IPR = P98 - P2
lower_bound = P2 - 1.5 * IPR
upper_bound = P98 + 1.5 * IPR
train = train[(train['Age'] >= lower_bound) & (train['Age'] <= upper_bound)]

len(train)
```
