```python
# exercise 5 - Anxiety Levels Across Different Genders

# Tasks:
# Import necessary libraries.
# Use Seaborn to create a bar plot comparing anxiety levels across genders from the dataset df.
# Customize the plot with an appropriate color palette and add a title.
# Display the plot.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('student mental health.csv')
gender = df['Choose your gender']
anxiety = df['Do you have Anxiety?']
```

```python
bars = plt.bar(
         gender,
         anxiety,
         color=["red", "blue"]
)

plt.title("Anxiety by Gender")
plt.xlabel("Gender")
plt.ylabel("Anxiety Level")

plt.show()
```

```python
# Exercise 6 - Relationship Between Age and Panic Attacks

# Tasks:
# Import necessary libraries.
# Convert panic attack responses to numeric values (e.g., Yes=1, No=0).
# Use Seaborn’s scatterplot to create a scatter plot with ‘Age’ on the x-axis and numeric panic attack responses on the y-axis.
# Customize the plot to improve readability by adding labels, a title, and adjusting point styles.
# Display the plot.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('student mental health.csv')

df["Do you have Panic attack?"] = df["Do you have Panic attack?"].replace({
    "No": 0,
    "Yes": 1
})

pa = df['Do you have Panic attack?']
Age = df['Age']

plt.scatter(Age, pa)

plt.title("Relationship Between Age and Panic attacks")
plt.xlabel("Age")
plt.ylabel("Panic attack")

plt.show()
```
