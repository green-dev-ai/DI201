```python
# class exercise 1
# Create a line plot for the function y = x^2 for x ranging from 0 to 10. Add appropriate labels and a title.

x = range(1,11,1)
y = [x ** 2 for x in range(10)]
plt.plot(x,y)

plt.title("EX 1 - Square Numbers")
plt.xlabel("Number")
plt.ylabel("Square")
plt.legend()
plt.grid()


plt.show()
```

```python
#---  Exercise 2

categories = ["product a", "product b", "product c", "product d", "product e"]
sales = [150, 240, 300, 200, 225]

patterns = ["/", "\\", "x", ".", "-"]

bars = plt.bar(
         categories,
         sales,
         color=["red", "blue", "green", "orange", "yellow"]

 )

for bar, pattern in zip(bars, patterns):
     bar.set_hatch(pattern)

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```
